```python
import pandas as pd
```

```python
# Load 4-1 data
pretest1 = pd.read_csv('data/4-1_Pre-test.csv')
phi1 = pd.read_csv('data/4-1_Day1(phi).csv')
human1 = pd.read_csv('data/4-1_Day2(human).csv')
turbo1 = pd.read_csv('data/4-1_Day3(turbo).csv')
posttest1 = pd.read_csv('data/4-1_Post-test.csv')

# Load 4-2 data
pretest2 = pd.read_csv('data/4-2_Pre-test.csv')
turbo2 = pd.read_csv('data/4-2_Day1(turbo).csv')
phi2 = pd.read_csv('data/4-2_Day2(phi).csv')
human2 = pd.read_csv('data/4-2_Day3(human).csv')
posttest2 = pd.read_csv('data/4-2_Post-test.csv')

# Load 4-3 data
pretest3 = pd.read_csv('data/4-3_Pre-test.csv')
human3 = pd.read_csv('data/4-3_Day1(human).csv')
turbo3 = pd.read_csv('data/4-3_Day2(turbo).csv')
phi3 = pd.read_csv('data/4-3_Day3(phi).csv')
posttest3 = pd.read_csv('data/4-3_Post-test.csv')
```

# Data cleaning

```python
# Remove empty columns
pretest1 = pretest1.dropna(axis=1, how='all')
phi1 = phi1.dropna(axis=1, how='all')
human1 = human1.dropna(axis=1, how='all')
turbo1 = turbo1.dropna(axis=1, how='all')
posttest1 = posttest1.dropna(axis=1, how='all')

pretest2 = pretest2.dropna(axis=1, how='all')
turbo2 = turbo2.dropna(axis=1, how='all')
phi2 = phi2.dropna(axis=1, how='all')
human2 = human2.dropna(axis=1, how='all')
posttest2 = posttest2.dropna(axis=1, how='all')

pretest3 = pretest3.dropna(axis=1, how='all')
human3 = human3.dropna(axis=1, how='all')
turbo3 = turbo3.dropna(axis=1, how='all')
phi3 = phi3.dropna(axis=1, how='all')
posttest3 = posttest3.dropna(axis=1, how='all')
```

```python
# Check for missing data
print('4-1')
print(pretest1.isnull().sum())
print(phi1.isnull().sum())
print(human1.isnull().sum())
print(turbo1.isnull().sum())
print(posttest1.isnull().sum())
```

```python
# Replace missing data with mean of column
pretest1 = pretest1.fillna(round(pretest1['Q1'].mean()))
pretest1.Q1
posttest1 = posttest1.fillna(round(posttest1['Q3'].mean()))
posttest1.Q3
```

```python
# Check for missing data
print('4-2')
print(pretest2.isnull().sum())
print(turbo2.isnull().sum())
print(phi2.isnull().sum())
print(human2.isnull().sum())
print(posttest2.isnull().sum())
```

```python
# Replace missing data with mean of column
pretest2 = pretest2.fillna(round(pretest2['Q4'].mean()))

# Remove Gabriella's data from 4-2 notebooks
turbo2 = turbo2.drop(turbo2[turbo2['Student'] == 'Gabriella'].index)
phi2 = phi2.drop(phi2[phi2['Student'] == 'Gabriella'].index)
human2 = human2.drop(human2[human2['Student'] == 'Gabriella'].index)
```

```python
# Check for missing data
print('4-3')
print(pretest3.isnull().sum())
print(human3.isnull().sum())
print(turbo3.isnull().sum())
print(phi3.isnull().sum())
print(posttest3.isnull().sum())
```

```python
# In 4-3 there are many missing values because not all the students were present for all the days. I will remove the data of students
# who were not present for all the days.

# Chilekwa, Daniel, Xiaanan, Emma, Gabriel and William were not present for all the days
# Remove Gabriel and William from Day 1
human3 = human3.drop(human3[human3['Student'] == 'Gabriel'].index).reset_index(drop=True)
human3 = human3.drop(human3[human3['Student'] == 'William'].index).reset_index(drop=True)

# Remove Gabriel, William, Chilekwa and Daniel from Day 2
turbo3 = turbo3.drop(turbo3[turbo3['Student'] == 'Gabriel'].index).reset_index(drop=True)
turbo3 = turbo3.drop(turbo3[turbo3['Student'] == 'William'].index).reset_index(drop=True)
turbo3 = turbo3.drop(turbo3[turbo3['Student'] == 'Chilekwa'].index).reset_index(drop=True)
turbo3 = turbo3.drop(turbo3[turbo3['Student'] == 'Daniel'].index).reset_index(drop=True)

# Remove Gabriel, William, Chilekwa, Daniel, Xiaanian and Emma from Day 3
phi3 = phi3.drop(phi3[phi3['Student'] == 'Gabriel'].index).reset_index(drop=True)
phi3 = phi3.drop(phi3[phi3['Student'] == 'William'].index).reset_index(drop=True)
phi3 = phi3.drop(phi3[phi3['Student'] == 'Chilekwa'].index).reset_index(drop=True)
phi3 = phi3.drop(phi3[phi3['Student'] == 'Daniel'].index).reset_index(drop=True)
phi3 = phi3.drop(phi3[phi3['Student'] == 'Xiaanian'].index).reset_index(drop=True)
phi3 = phi3.drop(phi3[phi3['Student'] == 'Emma'].index).reset_index(drop=True)

# Remove rows that contain at least one missing value from Post-test
posttest3 = posttest3.dropna()
```

```python
print(human3.shape)
print(turbo3.shape)
print(phi3.shape)
print(posttest3.shape)
print(pretest3.shape)
```

```python
# Check if we got rid of all missing values

print('4-1')
print(pretest1.isnull().sum())
print(phi1.isnull().sum())
print(human1.isnull().sum())
print(turbo1.isnull().sum())
print(posttest1.isnull().sum())

print('4-2')
print(pretest2.isnull().sum())
print(turbo2.isnull().sum())
print(phi2.isnull().sum())
print(human2.isnull().sum())
print(posttest2.isnull().sum())

print('4-3')
print(pretest3.isnull().sum())
print(human3.isnull().sum())
print(turbo3.isnull().sum())
print(phi3.isnull().sum())
print(posttest3.isnull().sum())
```

We got rid of all the nan values, amazing !

```python
# Join 4-1 data into one dataframe
df1 = pd.merge(phi1, human1, on='Student', suffixes=('_phi', '_human'))
df1 = pd.merge(df1, turbo1, on='Student', suffixes=('', '_turbo'))
df1.rename(columns={"Q4":"Q4_human", 
                    "Nb correct answers": "Nb correct answers_turbo", 
                    "Q1": "Q1_turbo", 
                    "Q2": "Q2_turbo",
                    "Q3": "Q3_turbo"}, 
                    inplace=True)
print(df1)
```

```python
# Join 4-2 data into one dataframe
df2 = pd.merge(turbo2, phi2, on='Student', suffixes=('_turbo', '_phi'))
df2 = pd.merge(df2, human2, on='Student', suffixes=('', '_human'))
df2.rename(columns={"Q4":"Q4_phi", 
                    "Nb correct answers": "Nb correct answers_human", 
                    "Q1": "Q1_human", 
                    "Q2": "Q2_human",
                    "Q3": "Q3_human"}, 
                    inplace=True)
print(df2)
```

```python
# Join 4-3 data into one dataframe
df3 = pd.merge(human3, turbo3, on='Student', suffixes=('_human', '_turbo'))
df3 = pd.merge(df3, phi3, on='Student', suffixes=('', '_phi'))
df3.rename(columns={"Q4":"Q4_turbo", 
                    "Nb correct answers": "Nb correct answers_phi", 
                    "Q1": "Q1_phi", 
                    "Q2": "Q2_phi",
                    "Q3": "Q3_phi"}, 
                    inplace=True)
print(df3)
```

```python
# Save merged dataframes to csv
df1.to_csv('clean_data/4-1_merged.csv', index=False)
df2.to_csv('clean_data/4-2_merged.csv', index=False)
df3.to_csv('clean_data/4-3_merged.csv', index=False)
```

```python
# Save cleaned pretest and posttest data to csv
pretest1.to_csv('clean_data/4-1_pretest.csv', index=False)
pretest2.to_csv('clean_data/4-2_pretest.csv', index=False)
pretest3.to_csv('clean_data/4-3_pretest.csv', index=False)

posttest1.to_csv('clean_data/4-1_posttest.csv', index=False)
posttest2.to_csv('clean_data/4-2_posttest.csv', index=False)
posttest3.to_csv('clean_data/4-3_posttest.csv', index=False)
```
